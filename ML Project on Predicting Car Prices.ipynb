```python
#Importing variance inflation factor from statsmodel to find out variance between values and to find out which values we can drop
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variables = data_cleaned[['Mileage','Year','EngineV']]
```

```python
vif=pd.DataFrame()
```

```python
vif['VIF']=[variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif['Features']=variables.columns
vif
```

```python
#From the above table we found out that the Year value has VIF greater than 10 so we will drop it
data_no_MC=data_cleaned.drop('Year',axis=1)
```

```python
data_no_MC
```

```python
#Creating dummies of the data
data_dummies=pd.get_dummies(data_no_MC,drop_first=True)
```

```python
data_dummies.head()
```

```python
data_dummies.columns.values
```

```python
cols = ['Log Price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
#Rearranging the columns
data_preprocessed = data_dummies[cols]
data_preprocessed.head()
```

```python
#Now, We will use the train test strategy to train our model
targets=data_preprocessed['Log Price']

inputs=data_preprocessed.drop(columns='Log Price')
```

```python
#First of all we have to scale our inputs using Standard Scaler from sklearn.preprocessing
from sklearn.preprocessing import StandardScaler

```

```python
scaler=StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
#Then we will transform them by scaler.transform
inputs_scaled=scaler.transform(inputs)
```

```python
#Importing Train Test Split from sklearn
from sklearn.model_selection import train_test_split
```

```python
#Assigning data to respective vaiables
x_train,x_test, y_train, y_test=train_test_split(inputs_scaled,targets,test_size=0.2,random_state=365)
```

```python
#Loading model into  a variable
reg=LinearRegression()
```

```python
#Fitting the regression using our data
reg.fit(x_train,y_train)
```

```python
#Calculating the Y Hat value i.e predicting values using our trained data
y_hat=reg.predict(x_train)
```

```python
y_hat
```

```python
#Plotting a graph between our trained y data and y hat values
plt.scatter(y_train, y_hat)
plt.xlabel('y_train',size=20)
plt.ylabel('y_hat',size=20)

plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Plotting a graph of residuals
sns.displot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

```python
#Finding out R squared
reg.score(x_train,y_train)
```

```python
#Finding out the constant
reg.intercept_
```

```python
#Finding out coefficients
reg.coef_
```

```python
#Finding out the weights and biases
reg_summary=pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights']=reg.coef_
reg_summary
```

```python
#Finding out unique brands
data_cleaned['Brand'].unique()
```

```python
#Testing our trained model using test data
y_hat_test=reg.predict(x_test)
```

```python
#Plotting a graph inbetween our test data Y and our test values of y hat

plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#Creating a DataFrame of our predicted values but with the exp values of our y hat test data
df_pf=pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf
```

```python
#Adding the values of our target values with expotential fucntion
df_pf['Target']=np.exp(y_test)
df_pf
```

```python
#We found missing values because the indexes of the target values was not resetted so we have to first reset those values
y_test=y_test.reset_index(drop=True)
y_test
```

```python
df_pf['Target']=np.exp(y_test)
df_pf
```

```python
#Finding out the residual values and adding them to our data frame
df_pf['Residual'] = df_pf['Target'] - df_pf['Predictions']
```

```python
#Adding a column of difference % 
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
